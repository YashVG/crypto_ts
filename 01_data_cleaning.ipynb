```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data preparation

```python
file_path = 'data/btc_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data
print(data.head())

# Ensure the date column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the date as the index
data.set_index('Date', inplace=True)

# Sort the data by date
data = data.sort_index()

```

```python
# Plot the closing price
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Bitcoin Closing Price')
plt.title('Bitcoin Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

 Clear inference can be made as prices have risen drastically. Now need to split it by relevant years. Prior to that, we need to pick out significant years statistically and/or mathematically, and group them into categories.

**As per ChatGPT, the suggested visual segment split should be:**
- 2014-2016 (Early Years)
- 2017-2018 (First major peak)
- 2019 (Post-peak stabilization)
- 2020-2022 (High volatility period)
- 2023-2024 (Recent stable growth)

```python
#Counting NaN values in data
data.isna().sum()
```

```python
#Removing NaN vals from data
data = data.dropna()
```

```python
#Splitting data set into segments above
early_years_data = data['2014-01-01':'2016-12-31']
first_major_peak_data = data['2017-01-01':'2018-12-31']
post_stable_data = data['2019-01-01':'2019-12-31']
volatile_data = data['2020-01-01':'2022-12-31']
recent_data = data['2023-01-01':]
```

```python
#Exporting data to separate csv files
early_years_data.to_csv('data/early_years_data.csv')
first_major_peak_data.to_csv('data/first_major_peak_data.csv')
post_stable_data.to_csv('data/post_stable_data.csv')
volatile_data.to_csv('data/volatile_data.csv')
recent_data.to_csv('data/recent_data.csv')
```
